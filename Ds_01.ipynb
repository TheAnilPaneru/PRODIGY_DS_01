```python
# Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('/content/population.csv')
```

```python
# Display the first few rows of the DataFrame
df.head()
```

```python
# Extract the relevant columns for the year 2022 and remove any missing values
df_2022 = df[['Country Name', '2022']].dropna()
```

```python
# Create a bar chart for the population of the first 10 countries in 2022
df_2022.head(10).plot(kind='bar', x='Country Name', y='2022', color='skyblue', legend=False)
plt.title('Population of the first 10 countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a histogram to show the distribution of population sizes in 2022
df_2022['2022'].plot(kind='hist', bins=10, color='skyblue')
plt.title('Distribution of Population Sizes in 2022')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
